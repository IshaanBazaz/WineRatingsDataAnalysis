# Part 2. Text Analysis

## 2.1 extract high requency words in description

```python
import re
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.probability import *
from itertools import chain
#from tqdm import tqdm
import codecs
with open('stopwords.txt') as f:
    stop_words = f.read().splitlines()
stop_words = set(stop_words)
```

```python
import string
tokenizer = RegexpTokenizer(r"\w+(?:[-']\w+)?") 
hf=pd.DataFrame([df.description]).transpose()
hf['words'] = ''
hf['words_str'] = ''
for i in range (0,129971):
  hf['words'][i]=tokenizer.tokenize(str(hf['description'][i]))
  hf['words'][i]=[token.lower() for token in hf['words'][i] if token not in stop_words]

for i in range (0,129971):
  hf['words_str'][i] = str(" ".join(hf['words'][i]))

freq = pd.Series(' '.join(hf.words_str).split()).value_counts()
```

```python
from google.colab import files
freq = freq[freq>5000]
freq.reset_index()
freq.to_csv('HighFreq.txt', sep=' ', index=False)
files.download("HighFreq.txt")
```

## 2.2 Find key words for describing Shiraz using TF-IDF

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import re
sk=pd.DataFrame([df[df.variety =='Shiraz'].description]) 
sk.to_csv('sk.csv',index=False,header=False)
sk = pd.read_csv('sk.csv')
tf_idf= TfidfVectorizer(stop_words=stop_words, use_idf=True)
X = tf_idf.fit_transform(sk)
sk = pd.DataFrame(X.toarray(), columns=tf_idf.get_feature_names())
sk
```

```python
skey=pd.DataFrame(list(sk.loc[:, sk.gt(0.4).any()]))
skey.to_csv("Shirazkey.txt", sep=',')
files.download("Shirazkey.txt")
```
